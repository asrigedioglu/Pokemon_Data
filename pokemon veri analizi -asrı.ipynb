```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sbn

```

```python
pokemon = pd.read_csv("pokemon.csv")
```

```python
pokemon_df= pokemon.copy()
```

```python
pokemon_df
```

```python
pokemon_df.columns ##sutun adlarını gösterir.
```

```python
yeni_sutunadı = {'#':'#', 'Name':"AD", 'Type 1':'Tip1', 'Type 2':'Tip2', 'HP':"CAN", 
                 'Attack':"SALDIRI", 'Defense':"SAVUNMA", 'Sp. Atk':'Sp. Atk',
       'Sp. Def':'Sp. Def', 'Speed':"HIZ", 'Generation':"NESİL", 'Legendary':"EFSANE"}

pokemon_df=pokemon_df.rename(columns=yeni_sutunadı)
```

```python
pokemon_df.info()
```

```python
pokemon_df.isnull().sum() ##boş değerleri sayar.
```

```python
pokemon_df.fillna("Yok",inplace=True) ## boş değerlere değer atadık ve inplace ile kalıcı hale getirdik.
```

```python
pokemon_df.isnull().sum() #görüldüğü gibi boş değerler "Yok" olarak değişti. Boş değerler 0 
```

```python
pokemon_df["Tip2"] 
```

```python
pokemon_df.drop("#",axis=1,inplace=True) ## "#" sutununu sildik.
```

```python
pokemon_df.reset_index() #index'i silme
```

```python
pokemon_df = pokemon_df.set_index("AD",inplace=True) #herhangi bir sutunu index olarak atama.
```

```python
pokemon_df
```

```python
tur_sayısı=pokemon_df["Tip1"].value_counts()
```

```python
tur_sayısı
```

```python
tur_adları = pokemon_df["Tip1"].value_counts().keys()
```

```python
tur_adları
```

Datayı incelemeye hazır hale getirdik. Bir kaç grafikle datayı analiz edelim.

```python
pokemon_figur = plt.figure().add_axes([0.5,0.5,2,1]) #figure oluşturup axis verdik. Yani tablo oluşturup alan oluşturduk.
pokemon_figur.plot(tur_adları,tur_sayısı,"r*-")
pokemon_figur.set_xlabel("Tür Adları")
pokemon_figur.set_ylabel("Tür Sayısı")
```

```python
pokemon_figur_barplot = plt.figure(figsize=(15,3)) #figure oluşturup axis verdik. Yani tablo oluşturup alan oluşturduk.
pokemon_figur_barplot= plt.title("Pokemon Türleri Barplot")
sbn.barplot(x=tur_adları ,y=tur_sayısı)
```

# TİP1'E GÖRE TOPLAM CAN 

```python
toplamcan_ort = pokemon_df.groupby("Tip1").describe()["CAN","mean"]
```

```python
toplamcan_ort.sort_values(ascending=False)
```

```python
pokemon_pie = plt.figure(figsize=(10,10)) ##grafik boyutu belirledik.
plt.pie(toplamcan_ort , labels=tur_adları , autopct="%1.1f%%", pctdistance=0.8, ) ##pie grafiği



# autopct= '.2f %%' , tur_sayisina göre yüzdelik dilimleri yazdık.
# pctdistance= 0.9 , grafikteki yüzde değerleri arasındaki mesafeyi ayarladık. 
# shadow=True, grafik altına gölge ekleme
# labeldistance: Dışarıda yer alan başlıkların grafiğe olan uzaklığını ayarlamamızı sağlar.
```

```python
pokemon_df
```

```python
tur_hızı = pokemon_df.groupby("Tip1")["HIZ"].mean()
## tip1'e göre gruplanan verilerin hızlarının ortalama değeri alındı. 
```

```python
plt.figure(figsize=(30,10))
plt.bar(tur_adları,tur_hızı, color="m")
```
